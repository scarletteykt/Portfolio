Странно, что не всегда совпадает возраст и пол. Скорее всего это связано с ошибками ввода, ошибкой выгрузки или некорректные данные. Будем удалять записи с клиентами, у которых есть противоречия в возрасте и в поле, чтобы в таблице остались только те записи, которые соответствуют критериям согласованности данных. 

```python
# Группируем по user_id и проверяем уникальность age
inconsistent_users = bank_scrooge.groupby('user_id').filter(
    lambda x: x['age'].nunique() > 1 or x['gender'].nunique() > 1
)['user_id']

# Выводим количество строк с одинаковым user_id и разным возрастом
print(f"Количество удаленных записей с одинаковым user_id и разными возрастом и полом: {len(inconsistent_users)}")

# Удаляем строки с одинаковым user_id и разным возрастом
bank_scrooge = bank_scrooge[~bank_scrooge['user_id'].isin(inconsistent_users)]

# Проверим результат
duplicates_by_user_id_after_clean = bank_scrooge[bank_scrooge.duplicated(subset=['user_id'], keep=False)]
print(f"Количество записей с одинаковым user_id после очистки данных: {len(duplicates_by_user_id_after_clean)}")
print("Строки с дублированными user_id:")
display(duplicates_by_user_id_after_clean.sort_values(by='user_id'))
```

У большей части дублированных клиентов пропущен возраст. Заполним пропущенные значения в столбце `age`, используя имеющийся возраст.

```python
# Заполняем значения nan в столбце age на основе среднего значения возраста для одинаковых user_id
bank_scrooge['age'] = bank_scrooge.groupby('user_id')['age'].transform(lambda x: x.fillna(x.mean()))

# Проверим результат
print("Строки с дублированными user_id:")
display(duplicates_by_user_id_after_clean.sort_values(by='user_id'))
```

**2.3. Проверим и обработаем пропуски.**

```python
# Проверим на наличие пропусков
missing_values_count = bank_scrooge.isna().sum()
missing_values_share = round(bank_scrooge.isna().mean() * 100, 2)

missing_data = pd.DataFrame({'missing_values_count': missing_values_count, 'missing_values_share': missing_values_share}).sort_values(by='missing_values_count', ascending=False)
missing_data
```

Пропущено 23% всех значений в столбце `balance` и 0.03% в столбце `age`.

Прежде чем обрабатывать пропуски, важно понять их природу. 
1. Определим тип пропусков: MCAR (Missing Completely at Random), MAR (Missing at Random), MNAR (Missing Not at Random).
2. Создадим для каждого столбца с пропусками бинарный признак. 1 — значение отсутствует. 0 — значение присутствует.
3. Построим сводную таблицу, где для каждой группы с пропусками и без рассчитаем среднее и медиану для других переменных.
4. Интерпретируем результаты. Если разница между группами минимальна, пропуски относятся к MCAR. Если различия значительны, то это скорее всего MAR или MNAR.

```python
# Создаем новые признаки для пропусков
missing_columns = ['age', 'balance']
for col in missing_columns:
    bank_scrooge[f'is_missing_{col}'] = bank_scrooge[col].isna().astype(int)

# Построим сводную таблицу с дополнительными признаками
pivot_tables = {}
additional_features = ['score', 'equity', 'est_salary', 'products', 'credit_card', 'churn']

for col in missing_columns:
    pivot_table = bank_scrooge.groupby(f'is_missing_{col}').agg({
        'balance': ['mean', 'median'],
        'age': ['mean', 'median'],
        **{feature: ['mean', 'median'] for feature in additional_features}
    }).reset_index()

    # Упростим названия столбцов
    pivot_table.columns = [f'is_missing_{col}'] + [
        f"{feature}_{agg}" for feature in ['balance', 'age'] + additional_features for agg in ['mean', 'median']
    ]
    pivot_tables[col] = pivot_table
    
# Выводим сводные таблицы для анализа
for _, pivot_table in pivot_tables.items():
    display(pivot_table)
```

Пропуски в возрасте нельзя считать случайными, похоже на MAR. Они связаны с клиентами с низким балансом, зарплатой и более высоким уровнем оттока. Но их всего 0.03%, поэтому удалим их.

```python
# Удалим пропуски в возрасте
bank_scrooge = bank_scrooge.dropna(subset=['age'])

# Проверим результат
print(f"Количество пропусков в столбце age: {bank_scrooge['age'].isna().sum()}")
```

Пропуски в balance также не случайны, похоже на MNAR. Они характерны для клиентов с высоким доходом, низкой оценкой собственности клиента и крайне низкой вероятностью оттока.

```python
# Данные для обучения модели
train_data = bank_scrooge[bank_scrooge['balance'].notna()]
test_data = bank_scrooge[bank_scrooge['balance'].isna()]

X_train = train_data[['est_salary', 'equity', 'churn', 'score', 'products', 'credit_card']]
y_train = train_data['balance']
X_test = test_data[['est_salary', 'equity', 'churn', 'score', 'products', 'credit_card']]

# Обучим модель
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# Предсказание пропусков
predicted_balance = model.predict(X_test)

# Заменим пропуски
bank_scrooge.loc[bank_scrooge['balance'].isna(), 'balance'] = predicted_balance

# Проверим результат
print(f"Количество пропусков в столбце balance: {bank_scrooge['balance'].isna().sum()}")
```

Выводы:
1. Пропуски в balance (23% от общего числа строк) относятся к MNAR (Missing Not At Random), так как они связаны с клиентами:
    - с высокой зарплатой (`est_salary`),
    - с низкой оценкой собственности (`equity`),
    - с низкой вероятностью оттока (`churn`).

Пропуски обработаны с использованием модели RandomForestRegressor на доступных данных с использованием признаков: `est_salary`, `equity`, `churn`, `score`, `products`, `credit_card`. Пропущенные значения предсказаны и заполнены.  

2.  Пропуски в age (0.03% от общего числа строк) относятся к MAR (Missing At Random), так как они связаны с клиентами:
    - с низким балансом (`balance`),
    - с низким доходом (`est_salary`),
    - с более высоким уровнем оттока (`churn`).

Учитывая незначительное количество пропусков (всего 0.03%), принято решение удалить строки с пропусками. Это минимально повлияет на данные, сохраняя их репрезентативность и упрощая анализ.

Пропуски обработаны с использованием модели RandomForestRegressor на доступных данных с использованием признаков: `est_salary`, `equity`, `churn`, `score`, `products`, `credit_card`. Пропущенные значения предсказаны и заполнены.

    - Пропуски в age (0.03% от общего числа строк) относятся к MAR (Missing At Random), так как они связаны с клиентами:
        - с низким балансом (`balance`),
        - с низким доходом (`est_salary`),
        - с более высоким уровнем оттока (`churn`).
Учитывая незначительное количество пропусков (всего 0.03%), принято решение удалить строки с пропусками. Это минимально повлияет на данные, сохраняя их репрезентативность и упрощая анализ.


4. Проверка и обработка некорректных данных:
    - Обнаружена и удалена 1 запись, где клиент имеет 0 продуктов.

**3.1. Сравнение портретов типичных клиентов, склонных и не склонных к оттоку.**

- Задача: выделить ключевые различия между клиентами, склонными (churn=1) и не склонными (churn=0) к оттоку.
- Методы:
    - рассчитаем средние и медианные значения для непрерывных признаков (balance, est_salary, score).
    - рассчитаем медиану для дискретных признаков (age, equity, products).
    - рассчитаем моду для категориальных признаков (city, gender).
    - построим таблицу с мерами центральной тенденции по группам.

```python
# Непрерывные признаки
continuous_features = ['age', 'balance', 'est_salary', 'score']

# Дискретные признаки
discrete_features = ['equity', 'products']

# Бинарные признаки
binary_features = ['credit_card', 'last_activity']

# Категориальные признаки
categorical_features = ['city', 'gender']
```

```python
# Строим таблицу с мерами центральной тенденции признаков для клиентов, склонных и не склонных к оттоку
summary_table = bank_scrooge.groupby('churn').agg({
    'age': ['mean', 'median'],
    'balance': ['mean', 'median'],
    'est_salary': ['mean', 'median'],
    'score': ['mean', 'median'],
    'equity': 'median',
    'products': 'median'
})
    
# Вычисляем моду для 'city' и 'gender' отдельно
mode_city = bank_scrooge.groupby('churn')['city'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)
mode_gender = bank_scrooge.groupby('churn')['gender'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)

# Добавляем в таблицу
summary_table['city_mode'] = mode_city
summary_table['gender_mode'] = mode_gender

# Транспонируем таблицу
transposed_summary_table = summary_table.T
transposed_summary_table.columns = ['Остались', 'Ушли']

# Переносим индекс в столбец и переименуем показатели
transposed_summary_table_sorted = transposed_summary_table.reset_index()
transposed_summary_table_sorted.rename(columns={
    'level_0': 'Показатель',
    'level_1': 'Метрика'
}, inplace=True)

# Вычисление разницы и разницы в процентах
transposed_summary_table_sorted['Разница'] = transposed_summary_table_sorted.apply(
    lambda row: row['Ушли'] - row['Остались'] if isinstance(row['Остались'], (int, float)) else "-",
    axis=1
)

transposed_summary_table_sorted['Разница (%)'] = transposed_summary_table_sorted.apply(
    lambda row: f"{((row['Ушли'] - row['Остались']) / row['Остались'] * 100):+.1f}%" 
    if isinstance(row['Остались'], (int, float)) and row['Остались'] != 0 else "-",
    axis=1
)

# Форматирование числовых данных
for col in ['Остались', 'Ушли', 'Разница']:
    transposed_summary_table_sorted[col] = transposed_summary_table_sorted[col].apply(
        lambda x: f"{x:,.1f}" if isinstance(x, (int, float)) else x
)
    
# Функция для добавления стилей к строкам с разницей больше 20%
def highlight_rows(data):
    def style_row(row):
        # Проверяем, есть ли разница в процентах и больше ли она 20%
        if isinstance(row['Разница (%)'], str) and '+' in row['Разница (%)']:
            diff_percent = float(row['Разница (%)'].replace('%', '').replace('+', ''))
            if diff_percent > 20:
                return ['background-color: #ffcccc'] * len(row)
        return [''] * len(row)
    return data.style.apply(style_row, axis=1)

# Применяем стили к таблице
styled_table = highlight_rows(transposed_summary_table_sorted)

# Отображаем таблицу с примененными стилями
display(styled_table)
```

Выводы:

    - Доход: различия минимальны. Доход не является значимым фактором, влияющим на отток.

```python
# Функция для построения графиков
def plot_distribution(data, feature, hue, limit=None):
    plt.figure(figsize=(10, 6))
    
    # Если задано ограничение, фильтруем данные
    if limit is not None:
        data = data[data[feature] <= limit]
    
    # Построение графика
    sns.histplot(data=data, x=feature, hue=hue, stat='density', common_norm=False, alpha=0.6)
    
    # Подписи легенды
    plt.legend(labels=["Ушли", "Остались"], title="Статус клиента")
    
    plt.title(f"Распределение {feature} у клиентов склонных и не склонных к оттоку")
    plt.xlabel(feature)
    plt.ylabel("Плотность")
    
    # Устанавливаем лимит на ось X, если задано ограничение
    if limit is not None:
        plt.xlim(0, limit)
    
    plt.grid()
    plt.show()


# Ограничение для 'balance', чтобы уменьшить влияние выбросов
balance_limit = bank_scrooge['balance'].quantile(0.98)

# Ограничение для 'est_salary', чтобы уменьшить влияние выбросов
est_salary_limit = bank_scrooge['est_salary'].quantile(0.97)

# Построим графики для непрерывных признаков
for feature in continuous_features:
    if feature == 'balance':
        plot_distribution(bank_scrooge, feature, hue='churn', limit=balance_limit)
    elif feature == 'est_salary':
        plot_distribution(bank_scrooge, feature, hue='churn', limit=est_salary_limit)
    else:
        plot_distribution(bank_scrooge, feature, hue='churn')
```

```python
# Функция для построения графика уровня оттока по заданному признаку
def plot_churn_by_feature(data, feature, title, xlabel, ylabel="Средний уровень оттока"):
    # Сводная таблица для расчета среднего уровня оттока
    churn_data = data.groupby(feature).agg({'churn': ['mean', 'count']}).reset_index()
    churn_data.columns = [feature, 'Средний уровень оттока', 'Количество записей']  # Переименование столбцов
    
    # Выводим таблицу
    print(f"Сводная таблица по {feature}:")
    display(churn_data)
    
    # Ограничиваем ось Y
    max_churn = churn_data['Средний уровень оттока'].max()
    y_limit = min(1.0, max_churn + 0.05)
    
    # Строим график
    plt.figure(figsize=(8, 6))
    sns.barplot(data=churn_data, x=feature, y='Средний уровень оттока', errorbar=None)
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.ylim(0, y_limit)
    
    plt.grid(axis='y')
    plt.show()
```

```python
# Анализ оттока категориальных признаков
for category in categorical_features:
    plot_churn_by_feature(
        data=bank_scrooge,
        feature=category,
        title=f"Уровень оттока по категории: {category}",
        xlabel=category.capitalize()
    )

# Сводная таблица для комбинации категорий
category_churn_combined = bank_scrooge.groupby(categorical_features).agg({'churn': 'mean'}).reset_index()

# Вывод таблицы
print("Уровни оттока по комбинации категорий:")
display(category_churn_combined)
```

## Проверка статистический гипотез

- **Гипотеза №1:**
    - H0 (нулевая гипотеза): Средний доход клиентов, склонных к оттоку, не отличается от среднего дохода клиентов, не склонных к оттоку.
    - H1 (альтернативная гипотеза): Средний доход клиентов, склонных к оттоку, отличается от среднего дохода клиентов, не склонных к оттоку.
    - Уровень значимости (α): 0.05. Это означает, что вероятность ошибки первого рода (отклонение нулевой гипотезы, когда она верна) составляет 5%.

- План работ:
    - Подготовка данных: разделить данные на две группы (churn=0 и churn=1).
    - Проверка нормальности: провести тест Колмогорова-Смирнова для оценки нормальности распределения доходов в каждой группе. 
    - Проверка гипотез:
        - Применить t-критерий Стьюдента, если данные нормально распределены.
        - Применить U-критерий Манна-Уитни, если нет.
    - Выводы.

```python
# Разделим данные на две группы (churn=0 и churn=1)
group_0 = bank_scrooge[bank_scrooge['churn'] == 0]['est_salary']
group_1 = bank_scrooge[bank_scrooge['churn'] == 1]['est_salary']

# Проверим нормальность распределения доходов с помощью теста Колмогорова-Смирнова
ks_test_0 = kstest(group_0, 'norm', args=(group_0.mean(), group_0.std()))
ks_test_1 = kstest(group_1, 'norm', args=(group_1.mean(), group_1.std()))

print(f"Нормальность для churn=0 (KS-test): p-value={ks_test_0.pvalue}")
print(f"Нормальность для churn=1 (KS-test): p-value={ks_test_1.pvalue}")

# Проверим гипотезы
t_stat, t_p_value = ttest_ind(group_0, group_1, equal_var=False)
u_stat, u_p_value = mannwhitneyu(group_0, group_1, alternative='two-sided')

print(f"T-критерий Стьюдента: t_stat={round(t_stat, 2)}, p-value={round(t_p_value, 2)}")
print(f"U-критерий Манна-Уитни: u_stat={u_stat}, p-value={u_p_value}")
```

**Объяснение расхождений:**
1. Нарушение нормальности: t-критерий основывается на предположении нормального распределения, что не выполняется в данном случае. Это делает его результаты не корректными.

2. Разные подходы к оценке:
    - T-критерий оценивает различия в средних значениях, которые сильно подвержены влиянию выбросов.
    - U-критерий оценивает различия в распределении рангов (медиан), что делает его более устойчивым к выбросам.


В данном случае распределение доходов не является нормальным, поэтому доверяем U-критерию Манна-Уитни, который показывает статистически значимую разницу (p-value<0.05). Отток клиентов действительно связан с их доходами.

Выводы:
- Тест на нормальность: в обеих группах распределение доходов не является нормальным: p-value < 0.05.
- T-критерий Стьюдента: t_stat=0.18, p-value = 0.86. P-value > 0.05. Нулевая гипотеза о равенстве средних доходов не отвергается.
- U-критерий Манна-Уитни: u_stat=6825137, p-value=0.00069. P-value < 0.05. Нулевая гипотеза о равенстве медиан доходов отвергается. Это указывает на статистически значимую разницу между доходами групп.

- **Гипотеза №2:**
    - H0 (нулевая гипотеза): Клиенты с высоким балансом (`balance` > 700,000) не имеют большей склонности к оттоку.
    - H1 (альтернативная гипотеза): Клиенты с высоким балансом (`balance` > 700,000) имеют большую склонность к оттоку.
    - Уровень значимости (α): 0.05. Это означает, что вероятность ошибки первого рода (отклонение нулевой гипотезы, когда она верна) составляет 5%.

Почему важно проверить эту гипотезу:
- Анализ показал, что высокий баланс — один из ключевых факторов, связанных с оттоком.
- На графиках распределений виден рост оттока в диапазоне `balance` > 700,000.


Для проверки будем использовать тест пропорций (z-test). Он лучше подходит для бинарных переменных.

```python
# Создаем бинарный признак
bank_scrooge['high_balance'] = (bank_scrooge['balance'] > 700000).astype(int)

# Подсчитываем доли клиентов с оттоком в каждой группе
count = [
    bank_scrooge[(bank_scrooge['high_balance'] == 1) & (bank_scrooge['churn'] == 1)].shape[0],
    bank_scrooge[(bank_scrooge['high_balance'] == 0) & (bank_scrooge['churn'] == 1)].shape[0]
]
nobs = [
    bank_scrooge[bank_scrooge['high_balance'] == 1].shape[0],
    bank_scrooge[bank_scrooge['high_balance'] == 0].shape[0]
]

# Применяем z-test
z_stat, z_p_value = proportions_ztest(count, nobs, alternative='larger')

print(f"Тест пропорций: z_stat={round(z_stat, 2)}, p-value={round(z_p_value, 7)}")
```

Выводы:
- Z-тест: z_stat=19.09, p-value=0.0. P-value > 0.05. Нулевая гипотеза отвергается. Это значит, что клиенты с высоким балансом (`balance` > 700,000) чаще склонны к оттоку.

Выводы:
- Существуют статистически значимые различия в медианах доходов между клиентами, склонными и не склонными к оттоку.
- Клиенты с высоким балансом (balance > 700,000) чаще склонны к оттоку.
