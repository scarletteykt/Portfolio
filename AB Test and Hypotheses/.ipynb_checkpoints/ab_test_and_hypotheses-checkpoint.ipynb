**Данные для анализа:**

В нашем распоряжении три датасета: файл *hypothesis.csv* хранит 9 гипотез по увеличению выручки интернет-магазина, *orders.csv* — информацию о заказах и *visitors.csv* — информацию о визитах пользователей.

1. Структура файла *hypothesis.csv*:
- `Hypothesis` — краткое описание гипотезы;
- `Reach` — охват пользователей по 10-балльной шкале;
- `Impact` — влияние на пользователей по 10-балльной шкале;
- `Confidence` — уверенность в гипотезе по 10-балльной шкале;
- `Efforts` — затраты ресурсов на проверку гипотезы по 10-балльной шкале. Чем больше значение Efforts, тем дороже проверка гипотезы.

2. Структура файла *orders.csv*:
- `transactionId` — идентификатор заказа;
- `visitorId` — идентификатор пользователя, совершившего заказ;
- `date` — дата, когда был совершён заказ;
- `revenue` — выручка заказа;
- `group` — группа A/B-теста, в которую попал заказ.

3. Структура файла *visitors.csv*:
- `date` — дата;
- `group` — группа A/B-теста;
- `visitors` — количество пользователей в указанную дату в указанной группе A/B-теста.

## Приоритизация гипотез.

**1.1. Импортируем все библиотеки, необходимые для анализа.**

```python
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
```

**1.2. Считываем данные из csv-файла и сохраняем в датафрейм hypothesis.**

```python
try:
    hypothesis = pd.read_csv('/datasets/hypothesis.csv')
except:
    hypothesis = pd.read_csv('https://code.s3.yandex.net/datasets/hypothesis.csv')
```

**1.3. Посмотрим на данные.**

```python
display(hypothesis)
```

**1.4. Применим фреймворк ICE для приоритизации гипотез. Отсортируем их по убыванию приоритета.**

```python
# Приведем названия столбцов к нижнему регистру
hypothesis.columns = hypothesis.columns.str.lower()

# Установка ширины отображения текста в столбце 'hypothesis'
pd.set_option('max_colwidth', 200)

hypothesis['ICE'] = (hypothesis['impact'] * hypothesis['confidence']) / hypothesis['efforts']
hypothesis['ICE'] = hypothesis['ICE'].round(2)
hypothesis['ICE_rank'] = hypothesis['ICE'].rank(ascending=False).astype(int)

display(hypothesis[['hypothesis', 'ICE', 'ICE_rank']].sort_values(by='ICE', ascending=False))
```

**1.5. Применим фреймворк ICE для приоритизации гипотез. Отсортируем их по убыванию приоритета.**

```python
hypothesis['RICE'] = (hypothesis['reach'] * hypothesis['impact'] * hypothesis['confidence']) / hypothesis['efforts']
hypothesis['RICE'] = hypothesis['RICE'].round(2)
hypothesis['RICE_rank'] = hypothesis['RICE'].rank(ascending=False).astype(int)

display(hypothesis[['hypothesis', 'RICE', 'RICE_rank']].sort_values(by='RICE', ascending=False))
```

```python
hypothesis.sort_values(by='RICE', ascending=False)
```

**Выводы:**

При использовании ICE приоритеты основываются только на трех факторах: Impact, Confidence и Efforts, что делает акцент на значимость, уверенность и трудозатраты, не учитывая охват. В RICE, помимо этих факторов, учитывается также Reach (охват), который показывает, насколько широкая аудитория будет затронута изменениями. Это и привело к изменениям в приоритизации:

1. Добавить форму подписки на все основные страницы — поднялась с 3-го места по ICE на 1-е место по RICE. Это произошло благодаря высокому показателю Reach (Reach = 10).
2. Добавить блоки рекомендаций товаров на сайт — поднялась с 5-го на 2-е место. Важную роль сыграл охват (Reach = 8).
3. Добавить два новых канала привлечения трафика — опустилась с 1-го на 3-е место, так как показатель Reach был относительно невысоким (Reach = 3).
4. Показать на главной странице баннеры — не изменилось, оставаясь на 4-м месте в обоих фреймворках.
5. Запустить акцию, дающую скидку на день рождения — опустилась с 1-го на 5-е место, так как её охват (Reach = 1) был низким.
6. Изменить структуру категорий — поднялась с 8-го на 6-е место благодаря высокому охвату (Reach = 8), несмотря на низкую оценку Impact и Confidence.
7. Запустить собственную службу доставки — опустилась с 6-го на 7-е место из-за невысокого охвата (Reach = 2).
8. Добавить страницу отзывов клиентов о магазине — опустилась с 7-го на 8-е место.
9. Изменить цвет фона главной страницы — осталась на последнем месте в обоих фреймворках.


RICE более точный метод приоритизации, так как он учитывает охват аудитории. Поэтому оставляем приоритизацию гипотез по RICE как финальную. Исходя из этого, гипотезу "Добавить форму подписки на все основные страницы" будем тестировать первой. Это изменение обладает высоким охватом (Reach = 10) и первым местом в приоритизации по RICE.

### Анализ A/B-теста

**1. Построим график кумулятивной выручки по группам.**

Начнём с подготовки данных.

```python
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

visitorsACummulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA': 'sum'}
    ),
    axis=1,
)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

visitorsBCummulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB': 'sum'}
    ),
    axis=1,
)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']

ordersADaily = (
    orders[orders['group'] == 'A'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = [
    'date',
    'ordersCummulativeA',
    'revenueCummulativeA',
]

ordersBDaily = (
    orders[orders['group'] == 'B'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = [
    'date',
    'ordersCummulativeB',
    'revenueCummulativeB',
]

data = (
    ordersADaily.merge(
        ordersBDaily, left_on='date', right_on='date', how='left'
    )
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')
)

print(data.head(5))
```

```python
# Построение графика кумулятивной выручки
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['revenueCummulativeA'], label='Группа A', color='blue')
plt.plot(data['date'], data['revenueCummulativeB'], label='Группа B', color='orange')

plt.title('Кумулятивная выручка по группам A и B', fontsize=16)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Кумулятивная выручка', fontsize=12)
plt.legend()
plt.grid()

plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

Выводы:
- На протяжении всего периода кумулятивная выручка группы B превышает кумулятивную выручку группы A. 
- В последние дни августа (особенно 19 августа) наблюдается резкий рост кумулятивной выручки группы B. Это связано с резким ростом (в 1.5 раза) кумулятивного среднего чека группы В 19 августа.

**2. Построим график кумулятивного среднего чека по группам.**

```python
# Расчет кумулятивного среднего чека для группы A
data['averageCheckCummulativeA'] = data['revenueCummulativeA'] / data['ordersCummulativeA']

# Расчет кумулятивного среднего чека для группы B
data['averageCheckCummulativeB'] = data['revenueCummulativeB'] / data['ordersCummulativeB']

# Построение графика
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['averageCheckCummulativeA'], label='Кумулятивный средний чек группы A', color='blue')
plt.plot(data['date'], data['averageCheckCummulativeB'], label='Кумулятивный средний чек группы B', color='orange')

plt.title('Кумулятивный средний чек по группам A и B', fontsize=16)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Кумулятивный средний чек', fontsize=12)
plt.legend()
plt.grid()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Выводы:
- Кумулятивный средний чек группы A, несмотря на начальные колебания, постепенно растет. В конце периода кумулятивный средний чек стабилизировался на уровне 7000.
- Кумулятивный средний чек группы В тоже растет. До 19 августа превосходство группы B было незначительным. 19 августа произошел резкий скачок до 10000, затем кумулятивный средний чек стабилизировался на уровне 9000.

**3. Построим график относительного изменения кумулятивного среднего чека группы B к группе A.**

```python
# Расчет относительного изменения кумулятивного среднего чека группы B к группе A 
data['relativeChangeCheckCummulative'] = (data['averageCheckCummulativeB'] / data['averageCheckCummulativeA']) - 1

# Построение графика относительного изменения
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['relativeChangeCheckCummulative'] * 100, label='Относительное изменение кумулятивного среднего чека группы B к A', color='green')

plt.title('Относительное изменение кумулятивного среднего чека группы B к группе A', fontsize=16)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Относительное изменение (%)', fontsize=12)
plt.axhline(y=0, color='black', linewidth=1.5, linestyle='--')
plt.legend()
plt.grid()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Выводы:
- Группа B 1 августа и с 11 по 14 августа показывала меньший средний чек, но в остальные дни значительно опережала группу A, особенно после резкого скачка 19 августа. В конце периода разрыв между группами сохраняется, но стабилизируется на уровне около 25%.

**4. Построим график кумулятивного среднего количества заказов на посетителя по группам.**

**10. Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным.**

Посчитаем статистическую значимость различий в среднем числе заказовна посетителя между группами по "сырым" данным. 

**Нулевая гипотеза:** различий в среднем числе заказов между группами нет. 

**Альтернативная гипотеза:** различия в среднем числе заказов между группами есть. 

**Уровень значимости (alpha):** 0.05. 

Создадим переменные ordersByUsersA и ordersByUsersB. В них для пользователей, которые заказывали хотя бы 1 раз, укажем число совершённых заказов. 

```python
ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': 'nunique'})
    .sort_values(by='transactionId', ascending=False)
)
ordersByUsersA.columns = ['userId', 'orders']

ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': 'nunique'})
    .sort_values(by='transactionId', ascending=False)
)
ordersByUsersB.columns = ['userId', 'orders']  
```

Объявим переменные sampleA и sampleB, в которых пользователям из разных групп будет соответствовать количество заказов. Тем, кто ничего не заказал, будут соответствовать нули. Это нужно, чтобы подготовить выборки к проверке критерием Манна-Уитни.

```python
sampleA = pd.concat(
    [
        ordersByUsersA['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleB = pd.concat(
    [
        ordersByUsersB['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

Применим критерий Манна-Уитни и отформатируем p-value, округлив его до трёх знаков после запятой.

```python
print("{0:.3f}".format(stats.mannwhitneyu(sampleA, sampleB)[1]))
```

Выведем относительный прирост среднего числа заказов группы B: среднее число заказов группы B / среднее число заказов группы A - 1.  Округлим до трёх знаков после запятой.

```python
print("{0:.3f}".format(sampleB.mean() / sampleA.mean() - 1))
```

Выводы:
- P-value = 0.017, что меньше уровня значимости 0.05, поэтому мы отвергаем нулевую гипотезу.
- Это означает, что различия в среднем числе заказов на посетителя между группами являются статистически значимыми.
- Относительный прирост среднего числа заказов группы B равен 13.8%.

Теперь посчитаем статистическую значимость различий в среднем чеке заказа между группами по "сырым" данным. 

**Нулевая гипотеза:** различий в среднем чеке между группами нет.

**12. Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным.**

Примем за аномальных пользователей тех, кто совершил от 3 заказов или совершил заказ дороже 40 000 рублей. Так мы уберём 2-3% пользователей с наибольшим числом заказов и от 2% до 3% пользователей с дорогими заказами. Сделаем срезы пользователей с числом заказов больше и равно 3 — usersWithManyOrders и пользователей, совершивших заказы дороже 40 000 — usersWithExpensiveOrders. Объединим их в таблице abnormalUsers. 

```python
orders_threshold = 3
revenue_threshold = 40000

usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] >= orders_threshold]['userId'],
        ordersByUsersB[ordersByUsersB['orders'] >= orders_threshold]['userId'],
    ],
    axis=0,
)

usersWithExpensiveOrders = orders[orders['revenue'] > revenue_threshold]['visitorId']

abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)


print(abnormalUsers.head(5))
print()
print(abnormalUsers.shape[0])
```

Всего 45 аномальных пользователей.

Узнаем, как действия аномальных пользователей повлияли на результаты теста.  Посчитаем статистическую значимость различий в среднем числе заказовна посетителя между группами по "очищенным" данным. 

**Нулевая гипотеза:** различий в среднем числе заказов между группами нет. 

**Альтернативная гипотеза:** различия в среднем числе заказов между группами есть. 

**Уровень значимости (alpha):** 0.05. 

```python
sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

Применим критерий Манна-Уитни и отформатируем p-value.

```python
print('{0:.3f}'.format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
```

Выведем относительный прирост среднего числа заказа группы B.

```python
print('{0:.3f}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1))
```

Выводы:


Результаты по среднему количеству заказов практически не изменились. 
- P-value = 0.007, что намного меньше уровня значимости 0.05, что позволяет отвергнуть нулевую гипотезу.
- Это указывает на то, что различия в среднем числе заказов на посетителя между группами являются статистически значимыми.
- Относительный прирост среднего числа заказов группы B равен 18.8%.

**Нулевая гипотеза:** различий в среднем чеке между группами нет.

RICE более точный метод приоритизации, так как он учитывает охват аудитории. Это и привело к изменениям в приоритизации гипотез. Финальная приоритезация гипотез:

2. Добавить блоки рекомендаций товаров на сайт — поднялась с 5-го на 2-е место. Важную роль сыграл охват (Reach = 8).

4. Показать на главной странице баннеры — не изменилось, оставаясь на 4-м месте в обоих фреймворках.
5. Запустить акцию, дающую скидку на день рождения — опустилась с 1-го на 5-е место, так как её охват (Reach = 1) был низким.
6. Изменить структуру категорий — поднялась с 8-го на 6-е место благодаря высокому охвату (Reach = 8), несмотря на низкую оценку Impact и Confidence.
7. Запустить собственную службу доставки — опустилась с 6-го на 7-е место из-за невысокого охвата (Reach = 2).

Остановить тест и зафиксировать победу группы B:

Аномальные пользователи, совершившие более 3 заказов или заказы дороже 40 000 рублей, не оказали значительного влияния на результат. Различия в среднем числе заказов остались статистически значимыми после их исключения, что подтверждает устойчивость преимущества группы B.
