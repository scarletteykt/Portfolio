**Удаляем аномалии**

**Удаляем аномальные значения путем отсечения 1 и 99 перцентелей (метод Winsorization).**

```python
# Функция для Winsorization
def winsorize_series(series, lower_quantile=0.01, upper_quantile=0.99):
    lower_bound = series.quantile(lower_quantile)
    upper_bound = series.quantile(upper_quantile)
    return series.clip(lower=lower_bound, upper=upper_bound)

# Функция для расчета перцентилей и вывода результатов
def calculate_percentiles(data, columns, percentiles=[0.01, 0.99]):
    results = {}
    for column in columns:
        results[column] = data[column].quantile(percentiles)
    return results

# Применение Winsorization к столбцам
columns_to_winsorize = ['last_price', 'total_area', 'rooms', 'ceiling_height', 
                        'living_area', 'kitchen_area', 'days_exposition']

for column in columns_to_winsorize:
    data[column] = winsorize_series(data[column])

# Проверка результатов
percentiles_results = calculate_percentiles(data, columns_to_winsorize)
for column, percentiles in percentiles_results.items():
    print(f'{column} перцентили после Winsorization:')
    print(percentiles, '\n')
```

**Изучаем данные. Похожи на реальные данные, нет выбивающихся значений. Применяем к датасету.**

```python
# Функция для удаления аномальных значений
def remove_outliers(df, columns, lower_quantile=0.01, upper_quantile=0.99):
    for column in columns:
        lower_bound = df[column].quantile(lower_quantile)
        upper_bound = df[column].quantile(upper_quantile)
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Укажем столбцы для анализа
columns_to_clean = ['last_price', 'total_area', 'rooms', 
                    'living_area', 'kitchen_area', 'days_exposition']

# Вывод перцентилей до очистки
print("Перцентили до удаления аномалий:")
for column in columns_to_clean:
    print(f'{column} перцентили:')
    print(data[column].quantile([0.01, 0.99]), '\n')

# Удаление аномальных значений
cleaned_data = remove_outliers(data, columns_to_clean)

# Проверка размеров данных до и после очистки
print(f'Размер данных до очистки: {data.shape}')
print(f'Размер данных после очистки: {cleaned_data.shape}')
```

**Считаем коэффициент корреляции Пирсона для параметров объектов.**

```python
# Порог значимости для корреляции
threshold = 0.5

# Вычисляем корреляционную матрицу
numeric_data = cleaned_data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_data.corr()

# Фильтруем только значимые корреляции
filtered_matrix = correlation_matrix[(np.abs(correlation_matrix) > threshold) & (correlation_matrix != 1)]

# Визуализация тепловой карты значимых корреляций
plt.figure(figsize=(10, 8))
sns.heatmap(filtered_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Значимые корреляции")
plt.show()
```

Наибольшее влияние на цену недвижимости оказывают её общая площадь (коэффициент корреляции Пирсона - 0.75), жилая площадь (0.59), площадь кухни (0.51) и количество комнат (0.43). Эти параметры являются ключевыми при определении стоимости объекта.

```python
# Группировка данных по населённым пунктам и подсчёт числа объявлений
top_localities = cleaned_data['locality_name'].value_counts().head(10).index.tolist()

# Создание сводной таблицы средней цены квадратного метра и числа объявлений
pivot_table = cleaned_data.query('locality_name in @top_localities').pivot_table(
    index='locality_name',
    values='price_per_sqm',
    aggfunc=['count', 'mean']
)

# Переименование столбцов для удобства
pivot_table.columns = ['count_of_ads', 'mean_price_per_sqm']

# Вывод сводной таблицы
display(pivot_table)
```

**Населённые пункты с самой высокой и низкой стоимостью квадратного метра.**

```python
# Определение населённых пунктов с самой высокой и самой низкой стоимостью квадратного метра
highest_price_locality = pivot_table['mean_price_per_sqm'].idxmax()
lowest_price_locality = pivot_table['mean_price_per_sqm'].idxmin()

print("\nНаселённый пункт с самой высокой стоимостью кв м:", highest_price_locality)
print("Населённый пункт с самой низкой стоимостью кв м:", lowest_price_locality)
```

**4.5. Определяем факторы, которые больше всего влияют на общую стоимость объекта.**

**Узнаем  зависит ли стоимость объектов в Санкт-Петербурге от расстояния до центра города.**

```python
spb_data = cleaned_data[cleaned_data['locality_name'] == 'санкт-петербург'].copy()

# Итерация по уникальным значениям километража и обновление информации по медианным ценам за кв м
for km in spb_data['cityCenters_nearest'].unique():
    median_price_per_sqm = spb_data.loc[spb_data['cityCenters_nearest'] == 'km', 'price_per_sqm'].median()
    spb_data.loc[spb_data['cityCenters_nearest'] == 'km', 'price_per_sqm'] = median_price_per_sqm

# Создание сводной таблицы средней цены за квадратный метр по расстоянию до центра
average_prices_data = spb_data.groupby('cityCenters_nearest')['price_per_sqm'].mean().reset_index()

# Построение графика изменения средней цены за квадратный метр в зависимости от расстояния до центра
plt.figure(figsize=(10, 6))
plt.plot(average_prices_data['cityCenters_nearest'], average_prices_data['price_per_sqm'], marker='o', linestyle='-')
plt.grid(True)
plt.show()
```

**Выводы:**
1. Чем дальше от центра, тем дешевле: стоимость за кв м снижается.
2. Есть локальные скачки. Это может быть связано с наличием элитных ЖК или развитой инфраструктурой, наличием зеленых зон.
